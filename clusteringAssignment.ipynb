```python
import pandas as pd

df = pd.read_csv("Citywide_Survey.csv")

df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
df.head()
```

```python
ecnoded_df = pd.get_dummies(df)
print('The new df is')
display(ecnoded_df.head(3))
print('The size of the dataframe is')
print(' We have', ecnoded_df.shape[0], 'rows, and', ecnoded_df.shape[1], 'columns')
```

```python
# استبدال القيم NaN بالوسط الحسابي لكل عمود
ecnoded_df.fillna(ecnoded_df.mean(), inplace=True)
```

```python
ecnoded_df.head()
```

```python
X = ecnoded_df.values
X
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# تطبيق K-Means بدون تطبيع
inertia = []
K_range = range(1, 11)  # اختبار القيم من 1 إلى 10 للمجموعات

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
# رسم Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of(K)')
plt.title('Elbow Method')
plt.show()
```

```python
from kneed import KneeLocator  # مكتبة لتحديد نقطة الكوع تلقائيًا

# إيجاد نقطة Elbow تلقائيًا
knee = KneeLocator(K_range, inertia, curve="convex", direction="decreasing")
optimal_k_before = knee.elbow
optimal_k_before
```

```python
kmeans_before = KMeans(n_clusters=optimal_k_before, random_state=42, n_init=10)
y_kmeans_before = kmeans_before.fit_predict(X)
```

```python
from sklearn.metrics import silhouette_score

# حساب Silhouette Score
silhouette_avg = silhouette_score(X, y_kmeans_before)
print(f"Silhouette Score بدون تطبيع: {silhouette_avg:.3f}")
```

```python
# Plot the clustered data points 
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans_before, cmap='viridis', marker='o', 
edgecolor='k', s=100) 
plt.scatter(kmeans_before.cluster_centers_[:, 0], kmeans_before.cluster_centers_[:, 
1], 
s=300, c='red', label='Centroids', edgecolor='k') 
plt.title(f'K-means Clustering (k={optimal_k_before})') 
plt.xlabel('Feature 1') 
plt.ylabel('Feature 2') 
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# تطبيق Min-Max Scaling
min_max_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(min_max_scaler.fit_transform(ecnoded_df), columns=ecnoded_df.columns)

# # تطبيق Standardization (Z-score)
# StandardScaler = StandardScaler()
# df_standard = pd.DataFrame(standard_scaler.fit_transform(ecnoded_df), columns=ecnoded_df.columns)

print("Original Data:\n", ecnoded_df)
print("\nMin-Max Scaled Data:\n", df_minmax)
```

```python
XX = df_minmax.values
XX
```

```python
# تطبيق K-Means بعد التطبيع
inertia_norm = []
K_range = range(1, 11)  # اختبار القيم من 1 إلى 10 للمجموعات

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(XX)
    inertia_norm.append(kmeans.inertia_)
```

```python
# رسم Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia_norm, marker='o', linestyle='--')
plt.xlabel('Number of (K)')
plt.title('Elbow Method')
plt.show()
```

```python
# إيجاد نقطة Elbow تلقائيًا بعد التطبيع
knee_norm = KneeLocator(K_range, inertia_norm, curve="convex", direction="decreasing")
optimal_k_after = knee_norm.elbow
optimal_k_after
```

```python
kmeans_after = KMeans(n_clusters=optimal_k_after, random_state=42, n_init=10)
y_kmeans_after = kmeans_after.fit_predict(XX)
```

```python
silhouette_avg_norm = silhouette_score(XX, y_kmeans_after)
print(f"Silhouette Score بعد التطبيع: {silhouette_avg_norm:.3f}")
```

```python
# Plot the clustered data points 
plt.scatter(XX[:, 0], XX[:, 1], c=y_kmeans_after, cmap='viridis', marker='o', 
edgecolor='k', s=100) 
plt.scatter(kmeans_after.cluster_centers_[:, 0], kmeans_after.cluster_centers_[:, 
1], 
s=300, c='red', label='Centroids', edgecolor='k') 
plt.title(f'K-means Clustering (k={optimal_k_after})')  
plt.legend()
```

```python
print("=== مقارنة النتائج ===")
print(f"عدد العناقيد الأمثل من Elbow Method قبل التطبيع: {optimal_k_before}")
print(f"عدد العناقيد الأمثل من Elbow Method بعد التطبيع: {optimal_k_after}")
print(f"Silhouette Score بدون تطبيع: {silhouette_avg:.3f}")
print(f"Silhouette Score بعد التطبيع: {silhouette_avg_norm:.3f}")

if silhouette_avg_norm > silhouette_avg:
    print("التطبيع حسّن من جودة التجميع.")
else:
    print("التطبيع جعله أسوأ.")
```
