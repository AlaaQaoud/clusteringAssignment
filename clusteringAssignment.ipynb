```python
ecnoded_df = pd.get_dummies(df)
print('The new df is')
display(ecnoded_df.head(3))
print('The size of the dataframe is')
print(' We have', ecnoded_df.shape[0], 'rows, and', ecnoded_df.shape[1], 'columns')
```

```python
# Ø§Ø³ØªØ¨Ø¯Ø§Ù„ Ø§Ù„Ù‚ÙŠÙ… NaN Ø¨Ø§Ù„ÙˆØ³Ø· Ø§Ù„Ø­Ø³Ø§Ø¨ÙŠ Ù„ÙƒÙ„ Ø¹Ù…ÙˆØ¯
ecnoded_df.fillna(ecnoded_df.mean(), inplace=True)
```

```python
ecnoded_df.head()
```

```python
X = ecnoded_df.values
X
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# ØªØ·Ø¨ÙŠÙ‚ K-Means Ø¨Ø¯ÙˆÙ† ØªØ·Ø¨ÙŠØ¹
inertia = []
K_range = range(1, 11)  # Ø§Ø®ØªØ¨Ø§Ø± Ø§Ù„Ù‚ÙŠÙ… Ù…Ù† 1 Ø¥Ù„Ù‰ 10 Ù„Ù„Ù…Ø¬Ù…ÙˆØ¹Ø§Øª

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
# Ø±Ø³Ù… Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of(K)')
plt.title('Elbow Method')
plt.show()
```

```python
from kneed import KneeLocator  # Ù…ÙƒØªØ¨Ø© Ù„ØªØ­Ø¯ÙŠØ¯ Ù†Ù‚Ø·Ø© Ø§Ù„ÙƒÙˆØ¹ ØªÙ„Ù‚Ø§Ø¦ÙŠÙ‹Ø§

# Ø¥ÙŠØ¬Ø§Ø¯ Ù†Ù‚Ø·Ø© Elbow ØªÙ„Ù‚Ø§Ø¦ÙŠÙ‹Ø§
knee = KneeLocator(K_range, inertia, curve="convex", direction="decreasing")
optimal_k_before = knee.elbow
optimal_k_before
```

```python
kmeans_before = KMeans(n_clusters=optimal_k_before, random_state=42, n_init=10)
y_kmeans_before = kmeans_before.fit_predict(X)
```

```python
from sklearn.metrics import silhouette_score

# Ø­Ø³Ø§Ø¨ Silhouette Score
silhouette_avg = silhouette_score(X, y_kmeans_before)
print(f"Silhouette Score Ø¨Ø¯ÙˆÙ† ØªØ·Ø¨ÙŠØ¹: {silhouette_avg:.3f}")
```

```python
# Plot the clustered data points 
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans_before, cmap='viridis', marker='o', 
edgecolor='k', s=100) 
plt.scatter(kmeans_before.cluster_centers_[:, 0], kmeans_before.cluster_centers_[:, 
1], 
s=300, c='red', label='Centroids', edgecolor='k') 
plt.title(f'K-means Clustering (k={optimal_k_before})') 
plt.xlabel('Feature 1') 
plt.ylabel('Feature 2') 
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# ØªØ·Ø¨ÙŠÙ‚ Min-Max Scaling
min_max_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(min_max_scaler.fit_transform(ecnoded_df), columns=ecnoded_df.columns)

# # ØªØ·Ø¨ÙŠÙ‚ Standardization (Z-score)
# StandardScaler = StandardScaler()
# df_standard = pd.DataFrame(standard_scaler.fit_transform(ecnoded_df), columns=ecnoded_df.columns)

print("Original Data:\n", ecnoded_df)
print("\nMin-Max Scaled Data:\n", df_minmax)
```

```python
XX = df_minmax.values
XX
```

```python
# ØªØ·Ø¨ÙŠÙ‚ K-Means Ø¨Ø¹Ø¯ Ø§Ù„ØªØ·Ø¨ÙŠØ¹
inertia_norm = []
K_range = range(1, 11)  # Ø§Ø®ØªØ¨Ø§Ø± Ø§Ù„Ù‚ÙŠÙ… Ù…Ù† 1 Ø¥Ù„Ù‰ 10 Ù„Ù„Ù…Ø¬Ù…ÙˆØ¹Ø§Øª

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(XX)
    inertia_norm.append(kmeans.inertia_)
```

```python
# Ø±Ø³Ù… Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia_norm, marker='o', linestyle='--')
plt.xlabel('Number of (K)')
plt.title('Elbow Method')
plt.show()
```

```python
# Ø¥ÙŠØ¬Ø§Ø¯ Ù†Ù‚Ø·Ø© Elbow ØªÙ„Ù‚Ø§Ø¦ÙŠÙ‹Ø§ Ø¨Ø¹Ø¯ Ø§Ù„ØªØ·Ø¨ÙŠØ¹
knee_norm = KneeLocator(K_range, inertia_norm, curve="convex", direction="decreasing")
optimal_k_after = knee_norm.elbow
optimal_k_after
```

```python
kmeans_after = KMeans(n_clusters=optimal_k_after, random_state=42, n_init=10)
y_kmeans_after = kmeans_after.fit_predict(XX)
```

```python
silhouette_avg_norm = silhouette_score(XX, y_kmeans_after)
print(f"Silhouette Score Ø¨Ø¹Ø¯ Ø§Ù„ØªØ·Ø¨ÙŠØ¹: {silhouette_avg_norm:.3f}")
```

```python
# Plot the clustered data points 
plt.scatter(XX[:, 0], XX[:, 1], c=y_kmeans_after, cmap='viridis', marker='o', 
edgecolor='k', s=100) 
plt.scatter(kmeans_after.cluster_centers_[:, 0], kmeans_after.cluster_centers_[:, 
1], 
s=300, c='red', label='Centroids', edgecolor='k') 
plt.title(f'K-means Clustering (k={optimal_k_after})')  
plt.legend()
```
